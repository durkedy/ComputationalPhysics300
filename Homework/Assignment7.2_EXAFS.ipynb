What is EXAFS?
- Extended X-ray Absorption Fine Structure (EXAFS), is the region in an X-ray Absorption spectrum located directly after the x-ray absorption edge.
- Observing and analysing this region of an X-ray absorption spectrum can reveal structural information about a species, such as information about interatomic distances in a molecule.
- X-ray absorption takes advantage of the photoelectric effect, except instead of visble or near-visible light, one uses photon energies typically on the order of keV. These photons are of sufficient energy to eject core electrons from an atom.
- The electrons ejected by the X-rays can be described by their wave properties; their wave functions will interact via scattering with adjacent atoms causing electron self-interference, which produces the EXAFS spectrum that we observe.

In my homework folder, I include jpg image titled "graphs1.jpg" which shows the region in energy space that I am talking about.

From the image, we see that when one collects an EXAFS spectrum, the graph shows the absorption as a function of energy.
However, we know from the de Broglie relation that photon energy is related to momentum, and thus in standard EXAFS analysis, the graphs that are worked with are plotted as a function of momentum, or "reciprocal space."
The y-axis as we shall see below, are calculated "chi(k)" values. Chi values for EXAFS are calculated from the EXAFS formula:

χ(E) = µ(E)−µ0(E) / ∆µ0(E)

For now, I won't go in to the theory behind the chi calculation (this is a huge sub-field), but the mu values are absorption coefficients for a particular species.

And the reciprocal values are in units of 1/Angstrom, and are calculated as such:

k =sqrt(2m(E −E0)/h**2)

Conclusively, EXAFS analysis is more sophisticated than what we have done here. However, these simple Fourier transforms we've performed
illuminate the powerful capability of Fourier transform on EXAFS data. EXAFS data that we collect at an X-ray source is initially plotted
as a function of energy of the incident X-rays. However, photon energy is directly correlated to momentum, and thus the plots that one works with
are chi(k) values in momentum, or reciprocal space. Since we are in reciprocal space, we can apply the Fourier transform to the chi(k) values to get back
a radial distribution function in real space. Finally, analysis is performed on this data to retrieve information about the system such as inter-atomic distances, which
can be very useful for crystal structure studies.
