```python
#import pyplot for plotting and numpy for functions.
import matplotlib.pyplot as plt
import numpy as np

#Declare variables in order to set the number of iterations for the mandelbrot function,
#and another to set the resolution of the image.
numIterations = 100
resolution = 1000

#Define the mandelbrot function to calculate values. I iterate over numIterations,
#which I can increase or decrease.
#The function contains an if statement - if after iterating over the mandelbrot function (z = z**2 + c)
#the modulus of z becomes greater than 2, then return the max number of iterations the function was able to achieve.
#otherwise, return 0. Values of 0 means that the complex number belongs to the mandelbrot set.
#I return x (the max number of iterations before abs(z)>2) to values that don't belong
#to the mandelbrot set to create an aesthetic gradient in the image.
def mandelbrot(c, numIterations):
    z = 0 + 0j
    
    for x in range(numIterations):
        z = z**2 + c
        if abs(z)>2:
            return x
    return 0

#Create two numpy arrays using linspace, with equally seperated values (based on the value of resolution variable)
#xgrid represents x values, y grid y values. Both are bounded between -2 and 2.
xgrid = np.linspace(-2, 2, resolution)
ygrid = np.linspace(-2, 2, resolution)

#Create variables width and height to keep track of values that I will store in a 2D array.
#Both are set equal to the resolution so they will have the same value.
width = resolution
height = resolution

#Create an empty 2D array that will store the values calculated by calling the function mandelbrot.
#Size is determined by width and height, based on the resolution that I set.
inmandelbrotset = np.empty((width, height))

#This series of four loops populates the rows and columns of the 2D array inmandelbrotset
#First, based on the resolution of my image grid (xgrid and ygrid), I determine complex values.
#Then, pass each complex value into the mandelbrot function until the array is populated.
for i_x in range(width):
    for i_y in range(height):
        
        cx = xgrid[i_x]
        cy = ygrid[i_y]
        c = complex(cx,cy)
        
        inmandelbrotset[i_y, i_x] = mandelbrot(c,numIterations)

#The following lines set figure size and create a plot based on the inmandelbroset array.
#I set the extent of the image between -2 and 2 on both axes.
#I experimented with some color maps, and chose this Spectral color map.
plt.figure(figsize=(100,100))
plt.matshow(inmandelbrotset, extent=(-2,2,-2,2), cmap="Spectral")
plt.show()
        
        
        
        
```
