```python
%matplotlib inline  

#importing relevant libraries for program
import numpy as np
import matplotlib.pyplot as plt
from math import log, exp

#Define the factorial function, which will calculate the factorial of a number.
def factorial(x):
    x = int(x)
    total = 1
    
    if (x >= 1):
        for i in range(2, x+1):
            total *= i
    return total

#Define function multiplicityA, to calculate the multiplicty for a given number of oscillators A, and energy n.
def multiplicityA(A,n):
    omegaA = exp(log(factorial(A))-log(factorial(n))-log(factorial(A-n)))
    return omegaA

#Define function multiplicityB, to calculate the multiplicity for a given number of oscillators B, and energy n.
def multiplicityB(B,n):
    omegaB = exp(log(factorial(B))-log(factorial(n))-log(factorial(B-n)))
    return omegaB

#Function Plot_OmegaTot defines an a array from 0 to 10 
def Plot_OmegaTot(A, B):
    
    #range of energy values q
    q = np.arange(0,101)
    omega = []

    #for loop to populate the array omega, which carries the total multiplicity values
    #for each combination of energy from systems A & B, where qtotal is always equal to 100.
    for n in range(0, 101):
        multA = multiplicityA(A, q[n])
        multB = multiplicityB(B, q[-n])
        omega.append(multA*multB)
        
    plt.plot(q, omega)
    plt.title('Na = ' + str(A)+ ', ' + 'Nb = ' + str(B))
    plt.xlabel('q')
    plt.ylabel('$\Omega(A)$')
    
    plt.show()
    
    
```

Below, I plot the total multiplicity Omega as a function of energy, q. The first plot uses the values of 300 and 200 for the number of oscillators in system Na and Nb, respectively. The second plot uses the values of 3000 and 2000. From the plots, one can see that the peak is centered at the same value of 60 - what changes is the amplitude of the peak (the multiplicity). This makes sense, since for a fixed energy, the total number of microstates (multiplicity) will increase by increasing Na and Nb, but will be centered around the same value governed by the total energy q.

```python
Plot_OmegaTot(300,200)
```

```python
Plot_OmegaTot(3000,2000)
```
